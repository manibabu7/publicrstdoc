## DataFrame Example

```
import pandas as pd

df = pd.DataFrame({"Year":[2020,2021,2022,2023],"Sales":[150,180,210,260]})
df
```

## Plot Example

```
import matplotlib.pyplot as plt
plt.plot(df["Year"], df["Sales"], marker="o")
plt.title("Sales Growth")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
